```python
import pandas as pd
import pandera as pa
from pandera import Column,Check
from sklearn.linear_model import LinearRegression
import joblib

schema=pa.DataFrameSchema({
"Units Sold": Column(int, Check.ge(0)),
"Unit Price": Column(float, Check.ge(0)),
"Sales Channel": Column(str, Check.isin(["Online", "Offline"]))
})

def validate(df:pd.DataFrame):
  schema.validate(df)

def load_training_data():
  return pd.DataFrame({
      "Units Sold":[10,20,30],
      "Unit Price":[100.0,200.0,300.0],
      "Sales Channel":["Online","Offline","Online"]
  })

def train_model(df:pd.DataFrame):
  X=df[["Units Sold","Unit Price"]]
  y=df["Units Sold"]*df["Unit Price"]
  model=LinearRegression()
  model.fit(X,y)
  return model

def save_model(model):
  joblib.dump(model,"model.pkl")

def training_pipeline():
  df=load_training_data()
  validate(df)
  model=train_model(df)
  save_model(model)
  print("Training completed successfully")


training_pipeline()



```

```python
def inference_pipeline(input_df):
 validate(input_df)
 model=joblib.load("model.pkl")
 predictions=model.predict(input_df[["Units Sold","Unit Price"]])
 return predictions
```

```python
 test_df=pd.DataFrame({
  "Units Sold":[5],
  "Unit Price":[100.0],
  "Sales Channel":["Online"]
})
 print(inference_pipeline(test_df))
```

```python
 bad_df=pd.DataFrame({
  "Units Sold":[-5],
  "Unit Price":[150.0],
  "Sales Channel":["Invalid"]
})
 print(inference_pipeline(bad_df))
```
